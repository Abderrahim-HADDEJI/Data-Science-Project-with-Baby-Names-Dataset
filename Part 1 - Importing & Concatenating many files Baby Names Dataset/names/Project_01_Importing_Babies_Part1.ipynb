```python
pd.read_csv("yob{}.txt".format(1880), header = None, names = ["name", "Sex", "Count"])
```

```python
years = list(range(1880, 2019))
years
```

```python
dataframes = []
for year in years:
    data = pd.read_csv("yob{}.txt".format(year), header =None, names = ["Name", "Sex", "Count"])
    dataframes.append(data)
```

```python
len(dataframes)
```

```python
pd.concat(dataframes, axis = 0, keys = years, names=["Year"]).droplevel(-1).reset_index()
```
